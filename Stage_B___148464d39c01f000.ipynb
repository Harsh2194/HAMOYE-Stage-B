Q12. From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the R^2 value in two d.p.?

```python
X_data = data[['T2']]
Y_data = data['T6']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_data, Y_data, test_size=0.3, random_state=1)

from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()

#fit the model to the training dataset
linear_model.fit(x_train, y_train)

#obtain predictions
predicted_values = linear_model.predict(x_test)

#r2_score
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 3)
```

Q13-Q20.

```python
data_new = data.drop(['date','lights'], axis = 1)
data_new.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_data = pd.DataFrame(scaler.fit_transform(data_new), columns=data_new.columns)
X_data = normalised_data.drop(columns=['Appliances'])
Y_data = normalised_data['Appliances']
```

Model generation

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_data, Y_data,
test_size=0.3, random_state=42)

from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
```

```python
#fit the model to the training dataset
linear_model.fit(x_train, y_train)

#obtain predictions
predicted_values = linear_model.predict(x_test)
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2) 
```

Residual Sum of Squares (RSS) 

```python
import numpy as np
```

```python
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2) 
```

Root Mean Square Error (RMSE)

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3) 
```

R-Squared 

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

Ridge Regression

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)

p1 = ridge_reg.predict(x_test)

from sklearn.metrics import mean_squared_error
rmse1 = np.sqrt(mean_squared_error(y_test, p1))
round(rmse1, 3)
```

Feature Selection and Lasso Regression

```python
from sys import ps2
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.0001)
lasso_reg.fit(x_train, y_train)

p2 = lasso_reg.predict(x_test)

from sklearn.metrics import mean_squared_error
rmse2 = np.sqrt(mean_squared_error(y_test, p2))
round(rmse2, 3)
```

```python
#comparing the effects of regularisation
def get_weights_data(model, feat, col_name):
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_data = pd.DataFrame(weights).reset_index()
  weights_data.columns = ['Features', col_name]
  weights_data[col_name].round(3)
  return weights_data
```

```python
linear_model_weights = get_weights_data(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_data = get_weights_data(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_data = get_weights_data(lasso_reg, x_train, 'Lasso_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_data, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_data, on='Features')
final_weights
```
